```python
import pandas as pd
import numpy as np
import os

# Data produk yang sudah ada
produk = [
    "Life Cat Can - Chicken", "Life Cat Can - Kitten Salmon", "Life Cat Can - Kitten Tuna",
    "Life Cat Can - Tuna", "Life Cat Pouch - Chicken Adult", "Life Cat Pouch - Chicken Tuna Adult",
    "Life Cat Pouch - Kitten Chicken", "Life Cat Pouch - Kitten Salmon", "Life Cat Pouch - Kitten Tuna",
    "Life Cat Pouch - Salmon Adult", "Life Cat Pouch Dus", "Life Cat Tofu - Macha", "Life Cat Tofu - Mango",
    "Trixie Cat Scratching Post", "Rogz Utility Dog Leash - Medium", "Drontal Cat Dewormer", 
    "Seresto Flea and Tick Collar for Dogs", "Kong Classic Dog Chew Toy", "Nutri-Vet Multivitamin Chewable",
    "Beaphar Ear Cleaner for Dogs & Cats", "Trixie Cat Laser Pointer", "Kong Puppy Teething Stick", 
    "EzyDog Chest Plate Harness for Dogs", "Frontline Plus Flea & Tick Treatment", "Trixie Cat Tunnel"
]

# Mapping produk ke kategori
produk_ke_kategori = {
    "Life Cat Can - Chicken": "Makanan",
    "Life Cat Can - Kitten Salmon": "Makanan",
    "Life Cat Can - Kitten Tuna": "Makanan",
    "Life Cat Can - Tuna": "Makanan",
    "Life Cat Pouch - Chicken Adult": "Makanan",
    "Life Cat Pouch - Chicken Tuna Adult": "Makanan",
    "Life Cat Pouch - Kitten Chicken": "Makanan",
    "Life Cat Pouch - Kitten Salmon": "Makanan",
    "Life Cat Pouch - Kitten Tuna": "Makanan",
    "Life Cat Pouch - Salmon Adult": "Makanan",
    "Life Cat Pouch Dus": "Makanan",
    "Life Cat Tofu - Macha": "Makanan",
    "Life Cat Tofu - Mango": "Makanan",
    "Trixie Cat Scratching Post": "Mainan",
    "Trixie Cat Laser Pointer": "Mainan",
    "Kong Puppy Teething Stick": "Mainan",
    "Kong Classic Dog Chew Toy": "Mainan",
    "Trixie Cat Tunnel": "Mainan",
    "Drontal Cat Dewormer": "Obat Hewan",
    "Seresto Flea and Tick Collar for Dogs": "Obat Hewan",
    "Nutri-Vet Multivitamin Chewable": "Obat Hewan",
    "Beaphar Ear Cleaner for Dogs & Cats": "Obat Hewan",
    "Frontline Plus Flea & Tick Treatment": "Obat Hewan",
    "Rogz Utility Dog Leash - Medium": "Aksesoris",
    "EzyDog Chest Plate Harness for Dogs": "Aksesoris"
}

# Metode Pembayaran
metode_pembayaran = ["Tunai", "Transfer", "QRIS"]

# Membuat transaksi data dummy
data_transaksi = []
transaksi_per_hari = {}

for i in range(1351):
    tanggal = pd.to_datetime(np.random.choice(pd.date_range("2024-01-01", "2024-12-10", freq="D"), 1)[0])
    num_items = np.random.randint(1, 4)  # Setiap transaksi membeli 1-3 barang
    items = np.random.choice(produk, num_items, replace=False)  # Memilih beberapa produk tanpa pengulangan
    metode_pembayaran_terpilih = np.random.choice(metode_pembayaran)
    
    total_harga_transaksi = 0
    transaksi_items = []  # Untuk menyimpan semua barang yang dibeli dalam satu transaksi

    # Menyimpan informasi produk yang dibeli
    for item in items:
        kategori_terpilih = produk_ke_kategori[item]
        jumlah_terjual = np.random.randint(1, 11)  # Batasan jumlah terjual maksimal 10
        harga_satuan = np.random.choice([12000, 4500, 20000, 13000, 30000, 25000])
        total_harga = harga_satuan * jumlah_terjual
        total_harga_transaksi += total_harga
        
        transaksi_items.append({
            "Nama Barang": item,
            "Kategori": kategori_terpilih,
            "Jumlah Terjual": jumlah_terjual,
            "Harga Satuan": harga_satuan,
            "Total Harga": total_harga
        })

    # Generate transaction ID
    tanggal_str = tanggal.strftime('%Y%m%d')
    if tanggal_str not in transaksi_per_hari:
        transaksi_per_hari[tanggal_str] = 1
    else:
        transaksi_per_hari[tanggal_str] += 1

    transaksi_id = f"TX-{tanggal_str}-{transaksi_per_hari[tanggal_str]:03d}"

    if total_harga_transaksi > 1000:  # Memastikan total harga > 1000
        # Gabungkan barang yang dibeli dalam satu transaksi
        nama_barang = ", ".join([item["Nama Barang"] for item in transaksi_items])
        kategori = ", ".join([item["Kategori"] for item in transaksi_items])
        jumlah_terjual = ", ".join([str(item["Jumlah Terjual"]) for item in transaksi_items])
        harga_satuan = ", ".join([str(item["Harga Satuan"]) for item in transaksi_items])
        total_harga = ", ".join([str(item["Total Harga"]) for item in transaksi_items])

        # Menambahkan transaksi ke dalam data
        data_transaksi.append([
            transaksi_id, tanggal, nama_barang, kategori, jumlah_terjual, harga_satuan, total_harga, metode_pembayaran_terpilih
        ])

# Membuat DataFrame
df = pd.DataFrame(data_transaksi, columns=[
    "ID Transaksi", "Tanggal Transaksi", "Nama Barang", "Kategori", "Jumlah Terjual", "Harga Satuan", "Total Harga", "Metode Pembayaran"
])

# Define the file path
file_path = r'D:\TB PAVD\data_transaksi_petshop.csv'

# Check if the directory exists
dir_path = os.path.dirname(file_path)
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

# Now save the DataFrame to the CSV file
df.to_csv(file_path, index=False)
```
